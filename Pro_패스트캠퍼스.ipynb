```
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```
title = []
main = []

url = 'https://search.naver.com/search.naver?date_from=20030520&date_option=6&date_to=20200916&dup_remove=1&nso=a%3At%2Cp%3A6m&post_blogurl=&post_blogurl_without=&query=%ED%8C%A8%EC%8A%A4%ED%8A%B8%EC%BA%A0%ED%8D%BC%EC%8A%A4&sm=tab_pge&srchby=title&st=sim&where=post&start={}'
for i in range(1,5000,10):
  URL = url.format(i)
  page = requests.get(URL)
  html = bs(page.text,'html.parser')
  content = html.find('ul',{'class':'type01'})
  blog = blog = content.find_all('li')
  for item in blog:
    title.append(item.find('dt').text)
    main.append(item.find('dd',{'class':'sh_blog_passage'}).text)
```

```
!pip install konlpy
from konlpy.tag import Okt
okt = Okt()
```

```
blog_title =[]
for item in title:
  item = item.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
  blog_title.append(item)
```

```
blog_main =[]
for item in main:
  item = item.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
  blog_main.append(item)
```

```
stopPos = ['Determiner','Conjunction','Exclamation','Josa','PreEomi','Eomi',
           'Suffix','Punctuation','Foreign','Alpha','Number','Unknown','KoreanParticle']
```

```
stopWord = ['곳','중','울','전','회','오다','제','차','하하','김종국','양세찬','등',
            '과','공','방','헛','확','하다','은','는','이','가','다','폐','로','비대','세',
            '정','당','문','시','기','콘','보','신','농','물','산','말','생','국','우아하다',
            '달','지','임','채다','레이','나선','명','비','대면','위','패스트','캠퍼스','린지','반','애','스','널리','우드','소식','소','싱','강','핫','개','일','저','원','아이','데이','션']
```

```
X = []
for item in blog_title:
  temp_X = []
  temp_X = okt.pos(item, stem=True)
  temp_X = [word[0] for word in temp_X if not word[1] in stopPos]
  temp_X = [word for word in temp_X if not word in stopWord]
  X.append(temp_X)
```

```
Y = []
for item in blog_main:
  temp_Y = []
  temp_Y = okt.pos(item, stem=True)
  temp_Y = [word[0] for word in temp_Y if not word[1] in stopPos]
  temp_Y = [word for word in temp_Y if not word in stopWord]
  Y.append(temp_Y)
```

```
title_list = []
for sentence in X:
  for token in sentence:
    title_list.append(token)
```

```
main_list = []
for sentence in Y:
  for token in sentence:
    main_list.append(token)
```

```
from collections import Counter
counting1 = Counter(title_list).most_common()
```

```
from collections import Counter
counting2 = Counter(main_list).most_common()
```

```
word1 = []
count_num1 = []
for item in counting1:
  word1.append(item[0])
  count_num1.append(item[1])
```

```
word2 = []
count_num2 = []
for item in counting2:
  word2.append(item[0])
  count_num2.append(item[1])
```

```
import pandas as pd
from pandas import DataFrame
data1 = {'word':word1,'counting':count_num1}
DF1 = DataFrame(data1)
DF1.head(40)
```
